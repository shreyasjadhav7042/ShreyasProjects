```python
df.columns = df.columns.str.title()
df.columns
```

```python
df.rename(
    columns = {
        'Primary_Skill': 'Skill',
        'Years_Of_Experience' : 'Experience',
        'Hourly_Rate (Usd)' : 'Hourly_Rate(Usd)'
    }, inplace = True
)
df.head(5)       
```

```python
df['Gender'] = df['Gender'].replace({
    'f': 'Female',
    'FEMALE': 'Female',
    'female': 'Female',
    'F': 'Female',
    'm': 'Male',
    'MALE': 'Male',
    'male': 'Male',
    'M': 'Male'
})
df.head(5)
```

```python
df.isna().sum()
```

```python
df.isna().mean()*100
```

```python
df['Age'].describe()
df['Age'].plot(kind='box')
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Age'].isna().sum()
```

```python
import math
df['Age'] = df['Age'].round()
```

```python
df['Experience'].plot(kind='box')
```

```python
df['Experience'] = df['Experience'].fillna(df['Experience'].median())
df['Experience'].isna().sum()
```

```python
df['Hourly_Rate(Usd)'] = df['Hourly_Rate(Usd)'].str.replace('USD','').str.replace('$','').astype(float)
df['Hourly_Rate(Usd)']
```

```python
df['Hourly_Rate(Usd)'] = df['Hourly_Rate(Usd)'].fillna(
    df.groupby('Skill')['Hourly_Rate(Usd)'].median()
)
df['Hourly_Rate(Usd)'].isna().sum()
```

```python
df['Rating'].plot(kind='box')
```

```python
df['Rating']=df['Rating'].fillna(df['Rating'].mean())
df['Rating'].isna().sum()
```

```python
df['Is_Active'] = df['Is_Active'].replace({
    'N' : False,
    '0' : False,
    'no' : False,
    'No' : False,
    'Y' : True,
    '1' : True,
    'yes' : True,
    'YES' : True,
    'Y' : True
})
df['Is_Active']
```

```python
df['Is_Active'] = df['Is_Active'].fillna('Unknown')
df['Is_Active'].isna().sum()
```

```python
df.loc[df['Client_Satisfaction']> 1, 'Client_Satisfaction'] /= 100
df['Client_Satisfaction']
```

```python
df['Client_Satisfaction'].plot(kind='box')
```

```python
df['Client_Satisfaction'] = df['Client_Satisfaction'].fillna(df['Client_Satisfaction'].mean())
df['Client_Satisfaction'].isna().sum()
```

```python
df.isna().sum()
```

```python
df['Rating'] = df['Rating'].round(1)
df['Client_Satisfaction'] = df['Client_Satisfaction'].round(2)
df
```
