```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df['Most_Used_Platform'].value_counts().plot(kind='bar')
plt.title('Most Used Social Media Platform')
plt.xlabel('Platform')
plt.ylabel('Number of Students')
plt.xticks(rotation=45)
```

```python
df['Affects_Academic_Performance'].value_counts().plot(kind='pie', autopct='%.2f%%', startangle=90)
plt.title('Impact of Social Media on Academic Performance')
```

```python
sns.histplot(x='Avg_Daily_Usage_Hours', data=df, kde=True)
plt.title('Distribution of Daily Social Media Usage')
```

```python
sns.boxplot(x='Most_Used_Platform', y='Avg_Daily_Usage_Hours', data=df)
plt.xticks(rotation=45)
plt.title('Daily Usage Across Social Media Platforms')
```

```python
df['Addiction_Level'] = pd.cut(df['Addicted_Score'],
                              bins=[0, 4, 7, 10],
                              labels=['Low', 'Medium', 'High'])

sns.countplot(x='Academic_Level', hue='Addiction_Level', data=df)
plt.title('Addiction Levels Across Academic Levels')
```

```python
df.groupby('Most_Used_Platform')['Addicted_Score'].mean().sort_values().plot(kind='barh')
plt.xlabel('Average Addiction Score')
plt.ylabel('Social Media Platform')
plt.title('Average Addiction Score by Platform')
```

```python
sns.scatterplot(x = 'Avg_Daily_Usage_Hours', y = 'Sleep_Hours_Per_Night', hue = 'Addicted_Score', data = df)
plt.title('Relationship Between Usage, Sleep, and Addiction')
```
